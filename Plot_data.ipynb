```python
import numpy as np
import matplotlib.pyplot as plt
import os
import sys
```

```python
timeseries_dict = {}

files = os.listdir('following_events_data')
csv_files = []
for file in files:
    if('csv' in file):csv_files.append(file)

for file in csv_files:
    file_path = os.path.join('following_events_data',file)
    data = np.loadtxt(file_path)
    timeseries_dict[file] = data

print('All following event data loaded.')
```

```python
plt.figure(figsize=[20,7])

for file in timeseries_dict:
    if('06_19' in file):
        plt.plot(timeseries_dict[file][:,5],timeseries_dict[file][:,1])
        
plt.ylabel('Speed [m/s]',fontsize=25)
plt.xlabel('GPS time [s]',fontsize=25)
plt.title('Data from 6/19/22',fontsize=30)
```

```python
plt.figure(figsize=[20,7])

for file in timeseries_dict:
    if('06_20' in file):
        plt.plot(timeseries_dict[file][:,5],timeseries_dict[file][:,1])
        
plt.ylabel('Speed [m/s]',fontsize=25)
plt.xlabel('GPS time [s]',fontsize=25)
plt.title('Data from 6/20/22',fontsize=30)
```

```python
files_with_stop_and_go = []

for file in timeseries_dict:
    if(np.min(timeseries_dict[file][:,1]) < 5.0):files_with_stop_and_go.append(file)

file = files_with_stop_and_go[0]
plt.figure(figsize=[20,25])
plt.subplot(3,1,1)
plt.scatter(timeseries_dict[file][:,5],timeseries_dict[file][:,1],c=timeseries_dict[file][:,3])
plt.ylabel('Speed [m/s]',fontsize=25)
plt.xlabel('GPS time [s]',fontsize=25)
plt.title('Data from 6/20/22',fontsize=30)
plt.colorbar()

plt.subplot(3,1,2)
plt.scatter(timeseries_dict[file][:,5],timeseries_dict[file][:,0],c=timeseries_dict[file][:,3])     
plt.ylabel('Spacing [m]',fontsize=25)
plt.xlabel('GPS time [s]',fontsize=25)
plt.title('Data from 6/20/22',fontsize=30)
plt.colorbar()

plt.subplot(3,1,3)
plt.scatter(timeseries_dict[file][:,5],timeseries_dict[file][:,2],c=timeseries_dict[file][:,3])  
plt.ylabel('Relative speed [m/s]',fontsize=25)
plt.xlabel('GPS time [s]',fontsize=25)
plt.title('Data from 6/20/22',fontsize=30)
plt.colorbar()
```

```python
plt.figure(figsize=[20,7])
plt.plot(timeseries_dict[file][:,5],timeseries_dict[file][:,3],label='Commanded',linewidth=3)
plt.plot(timeseries_dict[file][:,5],timeseries_dict[file][:,4],label='Realized',linewidth=3)

plt.ylabel('Acceleration [m/s^2]',fontsize=25)
plt.xlabel('GPS time [s]',fontsize=25)
plt.title('Data from 6/20/22',fontsize=30)
plt.legend(fontsize=25)
```
